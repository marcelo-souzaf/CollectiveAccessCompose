```python
translated = 0
no_change = 0

with open("messages.po", "r") as f:
    messages2 = f.read()
    with open("result.po", "w") as out:
        i = messages2.find("#: ")
        l = 0
        while True:
            i += messages2[i:].find("msgid") + 6
            if i == l + 5:
                break
            j = i + messages2[i:].find('"\nmsgstr') + 1
            k = j + 8
            # Writes everything before the new msgstr
            out.write(messages2[l:k])
            l = k + messages2[k:].find('"\n\n') + 1

            # If the sentence is translated, writes the translation
            if messages2[i:j] in sentences:
                if len(messages2[k:l].lstrip('"\n')) > 2:
                    no_change += 1
                    if sentences[messages2[i:j]] != messages2[k:l]:
                        print("Old: ", sentences[messages2[i:j]], "\nNew: ", messages2[k:l], sep="", end="\n\n")
                else:
                    translated += 1
                out.write(sentences[messages2[i:j]])
            # If not, writes the original msgstr
            else:
                out.write(messages2[k:l])

            i = l

print(no_change, translated)
```
